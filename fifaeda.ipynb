# Самые молодые как правило это полузащитники, возможно, это потому что они больше всех бегают

```python
def change1(x):
    if x is np.nan:
        return np.nan
    else:
        s = len(x)
        b = int(x[0:s-3])*0.45
        return b

def change2(x):
    if x is np.nan:
        return np.nan
    else:
        s1 = int(x[0])
        s2 = int(x[2])
        return s1*30+s2*3
```

```python
df.Weight = df.Weight.apply(change1)
df.Height = df.Height.apply(change2)
```

```python
pt = sns.pointplot(x=df["Kind of Pos"], y=df.Weight)
pt.set_title("Зависимость позиции игрока от его веса")
```

```python
pt = sns.pointplot(x=df.Weight, y=df.Position)
pt.set_title("Зависимость отдельных позиций игрока от его веса")
```

19. Есть ли зависимость между продолжительностью контракта игрока и его стоимостью? 

```python
def last(x):
    if (type(x)) == float:
        return None
    s = len(x)
    if s > 4:
        ans = int(x[s-4:s])
    else:
        ans = int(x)
    return ans
```

```python
df["Contract Valid Until"] = df["Contract Valid Until"].apply(last)
df.Joined = df.Joined.apply(last)
```

```python
df["Long"] = df["Contract Valid Until"] - df.Joined
```

```python
p = sns.lineplot(x=df.Long, y=df.Value)
p.set_title("Зависимость цены игрока от продолжительности контракта")
```

26. Насколько средняя зарплата (Wage) игроков с рейтингом (Overall) 90+ больше средней зарплаты с рейтингом [80 - 90]? Визуализируйте возможную зависимость

```python
w1 = df[df.Overall > 90]
w2 = df[(df.Overall >= 80) & (df.Overall <= 90)]
```

```python
mas_sum = [0]*2
mas_name = ["Overall > 90", "80 < Overall < 90"]
n1 = w1.shape[0]
n2 = w2.shape[0]

def summa1(x):
    global mas_sum
    mas_sum[0] += x
    
def summa2(x):
    global mas_sum
    mas_sum[1] += x
```

```python
w1.Poluchka.apply(summa1)
w2.Poluchka.apply(summa2)
mas_sum[0] /= n1
mas_sum[1] /= n2
```

```python
data = pd.DataFrame({
    'Wage_Mean1' : mas_sum,
    'Wage_Mean2' : mas_name
})
```

```python
explode =  (0,0.1,0,0  )
plt.pie(data.Wage_Mean1,  labels=data.Wage_Mean2, autopct='%1.1f%%')
plt.show()
```
